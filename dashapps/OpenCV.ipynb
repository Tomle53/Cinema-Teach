```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("../images/frame1.png")
gray_frame = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
blur_frame = cv.GaussianBlur(gray_frame, (17,17), 8)
```

```python
plt.imshow(blur_frame, cmap='gray', vmin=0, vmax=255)
```

```python
circles = cv.HoughCircles(blur_frame, cv.HOUGH_GRADIENT, 1.2, 100)

print(circles)
chosen = circles[0,:][0]
chosen = np.uint(np.around(chosen))
```

```python
cv.circle(img, (chosen[0], chosen[1]), 1, (255,0,0), 3)
cv.circle(img, (chosen[0], chosen[1]), chosen[2], (255,0,255), 3);
```

```python
plt.imshow(img)
```

```python
cv.imwrite("../images/frame1_out.png", img)
```
